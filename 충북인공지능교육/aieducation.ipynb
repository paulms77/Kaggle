순서형(0-9)

0: 손에 든 것은 없고, 인정받는 포커핸드가 아니다.
1: 한 쌍, 다섯 카드 내에서 한 쌍의 같은 순위
2: 2쌍, 5장의 카드 내에서 같은 순위의 2쌍
3: 3종류, 5종류 내 3종류
4: 직진; 5개의 카드, 공백 없이 순차적으로 순위가 매겨짐
5: 플러시, 같은 양복의 카드 5장
6: 풀 하우스, 페어 + 동종 3위
7: 4종류, 5종류 이내 4종류
8: 직진, 직진 + 플러시
9: 로열 플러시, {에이스, 킹, 퀸, 잭, 텐} + 플러시

```
train = train.drop("Id",axis=1)
test = test.drop("Id",axis=1)
```

```
y_train_df = train['CLASS']
x_train_df = train.drop("CLASS",axis=1)
```

```
x_train_df
```

```
import keras
```

```
y_train_df = keras.utils.to_categorical(y_train_df)
```

```
y_train_df.shape, x_train_df.shape
```

```
#########################################   DNN   ##############################################
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x_train_df, y_train_df, test_size=0.2, random_state=21)
```

```
from keras import models
from keras import layers
import tensorflow as tf
from keras.layers import BatchNormalization
from keras.layers import Dropout
from keras import optimizers
```

```
#model1.add(BatchNormalization()) # 정규화
#model1.add(Dropout(0.2)) # 과적합 방지
#model.add(Flatten()) # 인풋을 일렬화
#keras.layers.RepeatVector(n) # 인풋을 n번 반복

model1 = models.Sequential()
model1.add(layers.Dense(100, activation='relu',kernel_initializer='normal', input_shape=(10,))) #입력층 input_shape 10개
model1.add(layers.Dense(100, activation='relu',kernel_initializer='uniform'))
model1.add(layers.Dense(100, activation='relu',kernel_initializer='uniform'))
model1.add(layers.Dense(50, activation='relu',kernel_initializer='uniform'))
model1.add(layers.Dense(50, activation='relu',kernel_initializer='uniform'))
model1.add(layers.Dense(10, activation='relu',kernel_initializer='normal'))
model1.add(layers.Dense(9, activation='softmax'))
# glorot_normal
#random_normal
#normal
```

```
from keras import optimizers

# sgd = SGD(lr=0.001, momentum=0.9, nesterov=True)
model1.compile(optimizer=optimizers.Adam(lr=0.001), # AdamMax, SGD, RMSprop (lr=0.001)
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```
from keras.callbacks import EarlyStopping
from keras.callbacks import ReduceLROnPlateau

def Callbacks():
    erl = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='min', 
                        restore_best_weights=True)
    rdc = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, verbose=1, mode='min')
    return [erl,rdc]
```

```
history = model1.fit(x_train,
                    y_train,
                    epochs=1000, # 모델 학습 횟수
                    batch_size=48,# 한번에 몇개의 데이터를 학습할 것인가
                    #callbacks=Callbacks(),# earlystopping, reduce~
                    validation_split=0.2)
```

```
results = model1.evaluate(x_test, y_test)
print('Test accuracy: ', results[1])
```
